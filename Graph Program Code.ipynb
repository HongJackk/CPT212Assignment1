# Number of Primitive Operations vs Number of Input (Constant Length)

## Part 1 Graph

```python
import matplotlib.pyplot as plt

# Dataset A
A = [
    (2, 4551),
    (3, 4924),
    (4, 5293),
    (5, 5662),
    (6, 6037),
    (7, 6406),
    (8, 6775)
]

# Extract x (Number of Input) and y (Primitive Operations)
x = [p[0] for p in A]
y = [p[1] for p in A]

# Plot setup
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue')
plt.scatter(x, y, color='blue')

# Show the coordinate
for xi, yi in zip(x, y):
    plt.text(xi, yi + 90, f'({xi}, {yi})', ha='center', fontsize=9, color='black')

# Graph label
plt.xlabel('Number of Input', labelpad=15)
plt.ylabel('Number of Primitive Operation', labelpad=15)
plt.title('Number of Primitive Operations vs Number of Input (Constant Length - 3 Digit)')
plt.grid(True)
plt.tight_layout()

plt.show()

import numpy as np

x = np.array([2, 3, 4, 5, 6, 7, 8])
y = np.array([4551, 4924, 5293, 5662, 6037, 6406, 6775])

# Fit a line y = mx + c
m, c = np.polyfit(x, y, 1)
print(f"y = {m:.2f}x + {c:.2f}")
```
